```python
data = pd.read_csv("/content/archive (9) (1).zip")
data.head()
```

```python
sentiment_counts = data['Facebook'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="Blues_r")
plt.title('Sentiment Analysis of Twitter Data')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
