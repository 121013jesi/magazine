```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = "https://www.kaggle.com/c/titanic/data"
# Assuming you've already downloaded the dataset and placed it in the same directory as this script
data = pd.read_csv("/content/archive (12).zip")

# Data Exploration
print(data.head())
print(data.info())

# Summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Data Cleaning
# Handle missing values
# For example, we can fill missing values in the 'Age' column with the median age
data['Age'].fillna(data['Age'].median(), inplace=True)

# Explore relationships between variables
# Survival by Sex
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Survival by Sex')
plt.show()

# Survival by Pclass
sns.countplot(x='Survived', hue='Pclass', data=data)
plt.title('Survival by Pclass')
plt.show()

# Age distribution
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()

# Fare distribution
sns.histplot(data['Fare'], bins=20, kde=True)
plt.title('Fare Distribution')
plt.show()

# Identify patterns and trends
# Survival rate by Sex and Pclass
survival_rate = data.groupby(['Sex', 'Pclass'])['Survived'].mean().unstack()
print(survival_rate)

# Average fare by Pclass
average_fare = data.groupby('Pclass')['Fare'].mean()
print(average_fare)

# Survival rate by Embarked
survival_embarked = data.groupby('Embarked')['Survived'].mean()
print(survival_embarked)
```
