```python
data_fraud = df[df['Fraud'] == 1]
data_fraud.info()
```

```python
column_barchart1 = ['Time', 'Merchant Group', 'Country of Transaction', 'Shipping Address', 'Age', 'Bank', 'Country of Residence']

for column in column_barchart1:
    if data_fraud[column].dtype in ['int64', 'float64']:
        plt.figure(figsize=(11, 5))
        sns.histplot(data_fraud[column])
        plt.title(f'Distribution of {column}')
        plt.xlabel('Value')
        plt.ylabel('Frequency')

        plt.show()
```
